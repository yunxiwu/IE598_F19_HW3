```python
import pandas as pd
import numpy as np
import pylab
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\\Users\Wincy\Desktop\\HY_Universe_corporate bond.csv")
print(data.head)
print(data.tail())
print(data.shape)
```

```python
print(data.shape)
print(data.describe(percentiles=[0.25, 0.5,0.75]))
```

```python
print(data.info())
```

```python
#Quantile‐Quantile Plot
weekly_mean_volume = data['weekly_mean_volume']
stats.probplot(weekly_mean_volume, dist="norm", plot=plt)
plt.show()
n_trades = data['n_trades']
stats.probplot(n_trades, dist="norm", plot=plt)
plt.show()
```

```python
#swarm plot
_= sns.swarmplot('Months in Both','LIQ SCORE',data=data)
plt.show()

#histogram
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
_= plt.hist(data["LIQ SCORE"], bins=20)
_= plt.xlabel('Coupon Type')
_= plt.ylabel('LIQ SCORE')
plt.show()
```

```python
#ECDF
X = np.sort(data["weekly_mean_volume"])
y = np.arange(1, len(X)+1) / len(X) 
plt.plot(X, y, marker='.', linestyle='none')
plt.xlabel("weekly_mean_volume")
plt.ylabel("ECDF")
plt.show()
```

```python
#Parallel Coordinates Plots
from pandas.plotting import parallel_coordinates
subdata = data[['LIQ SCORE','Issued Amount','days_diff_max','n_trades','Coupon Type']][:100]
parallel_coordinates(subdata, 'Coupon Type')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.6,-0.1),  ncol = 2, fancybox = True, shadow = True)
plt.show()
```

```python
#Cross-plot
plt.scatter(data['n_trades'], data['percent_uncapped'])
plt.xlabel('n_trades')
plt.ylabel('percent_uncapped')
plt.show()
```

```python
#heat map
correlation_matrix = pd.DataFrame(data.corr())
sns.heatmap(correlation_matrix)
```
